```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

## Pipelines for Transformations

```python
numerical_features = X_train.select_dtypes(include = 'number').columns.tolist()
categorical_features = X_train.select_dtypes(exclude = 'number').columns.tolist()
```

```python
def fit_evaluate(method, CS = False, NZV = False, Corr = False, var_threshold = 0.01):    
    steps_num = []
    transformed = False
    
    if CS:
        steps_num.append(('scaleCenter', StandardScaler())) 
    if NZV:
        steps_num.append(('removeNZV', VarianceThreshold(threshold = var_threshold)))
    # if Corr:
        # steps.append(TBD)
    if CS or NZV or Corr:
        transformed = True
    
    process_num = Pipeline(steps_num)
    process_cat = Pipeline([('one-hot', OneHotEncoder(handle_unknown = 'ignore', sparse_output = False))])
    if transformed:
        process_all = [('number', process_num, numerical_features),
                       ('category', process_cat, categorical_features)]
    else:
        process_all = [('category', process_cat, categorical_features)]
    full_processor = ColumnTransformer(process_all)
    
    model = Pipeline([('preprocess', full_processor),
                      ('method', method)])
    model.fit(X_train, y_train)
    
    cv = LeaveOneOut()
    scores = cross_val_score(
        model, X, y,
        scoring='neg_mean_absolute_percentage_error',
        cv = cv,
        n_jobs=-1)
    loocv = -np.mean(scores)
    
    result = pd.DataFrame(
        {'model': str(method),
         'RMSE': np.sqrt(mean_squared_error(y_test, model.predict(X_test))),
         'Rsquared':r2_score(y_test, model.predict(X_test)),
         'MAE': mean_absolute_error(y_test, model.predict(X_test)),
         'MAPE': mean_absolute_percentage_error(y_test, model.predict(X_test)),
         'LOOCV_MAPE': loocv,
         'Transformed': transformed,
         'Scaled': CS,
         'NZV Removed': NZV,
         'MC Removed': Corr},
        index = [0])
    
    return result #, model
```

```python
results = pd.DataFrame() #columns = ['model','RMSE','Rsquared','MAE','MAPE'])
```

## OLS
CS, NZV, Corr

```python
ols_raw = fit_evaluate(LinearRegression())
ols_trans = fit_evaluate(LinearRegression(), CS = True, NZV = True, Corr = True)
results = pd.concat([results, ols_raw, ols_trans])
display(ols_raw, ols_trans)
```

## PLS
CS

```python
pls_raw = fit_evaluate(PLSRegression(n_components = 5))
pls_trans = fit_evaluate(PLSRegression(n_components = 5), CS = True)
results = pd.concat([results, pls_raw, pls_trans])
display(pls_raw, pls_trans)
```

### Ridge
CS, NZV

```python
ridge_raw = fit_evaluate(Ridge(alpha = 1.0))
ridge_trans = fit_evaluate(Ridge(alpha = 1.0), CS = True, NZV = True)
results = pd.concat([results, ridge_raw, ridge_trans])
display(ridge_raw, ridge_trans)
```

## LASSO
CS, NZV

```python
lasso_raw = fit_evaluate(linear_model.Lasso(alpha = 0.1))
lasso_trans = fit_evaluate(linear_model.Lasso(alpha = 0.1), CS = True, NZV = True)
results = pd.concat([results, lasso_raw, lasso_trans])
display(lasso_raw, lasso_trans)
```

CS

## Results

```python
results = results.reset_index(drop = True)

results = (results[
    ~(results.duplicated('model', keep = False)) |
    (results['Transformed'] != False)]
 .sort_values('MAPE')
 .reset_index(drop = True)
)
```

```python
results
```

```python
ranks = (results
 .assign(
     RMSE_rank = results['RMSE'].rank(),
     R2_rank = results['Rsquared'].rank(ascending = False),
     MAE_rank = results['MAE'].rank(),
     MAPE_rank = results['MAPE'].rank(),
     LOOCV_rank = results['LOOCV_MAPE'].rank())
 .filter(['model', 'RMSE_rank', 'R2_rank', 'MAE_rank', 'MAPE_rank', 'LOOCV_rank'], axis = 1)
 .assign(Avg_rank = lambda x: x.mean(axis = 1))
 .sort_values('Avg_rank')
)
```

```python
(results
 .drop(['Transformed','Scaled','NZV Removed','MC Removed'], axis = 1)
 .merge(ranks[['model','Avg_rank']], how = 'left')
)
```
