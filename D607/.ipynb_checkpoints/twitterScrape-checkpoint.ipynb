```python
candidates = pd.read_csv('data/candidates2022.csv')
```

```python
len(candidates.fec_id.unique())
```

```python
twitter_ids = candidates['twitter_id'].dropna()
twitter_ids = [*set(twitter_ids)]
```

```python
len(twitter_ids)
```

```python
%%capture

# Created a list to capture all missing twitter accounts
exceptions = []
names_from_ids = []

for user_id in twitter_ids:

    # Use ID to obtain username
    user_id = int(user_id)
    try:
        user_name = sntwitter.TwitterUserScraper(user_id).entity.username
    except:
        exceptions.append(user_id)
        continue
    
    # Add resulting user name to candidates df
    index = candidates.index[candidates['twitter_id'] == user_id]
    candidates.loc[index, 'user_name'] = user_name
    names_from_ids.append(user_name)
```

```python
print(
    f'Successes - Twitter user names succesfully scraped: {len(twitter_ids) - len(exceptions)}',
    f'\nExceptions - Twitter IDs not matched: {len(exceptions)}',
)
```

```python
user_names1 = [x for x in candidates['user_name'].dropna()]
user_names2 = [x for x in candidates['twitter_name'].dropna()]
```

```python
user_names = [*set(user_names1 + user_names2)]
```

```python
# Created a list to append all tweet attributes(data)
attributes_container = []
start = time.perf_counter()

# Using TwitterSearchScraper to scrape data and append tweets to list
for user_name in user_names:
    for tweet in sntwitter.TwitterSearchScraper(f'from:{user_name} since:2022-01-01 until:2023-01-01').get_items():
        attributes_container.append([tweet.user.id,
                                     tweet.user.username,
                                     tweet.user.verified,
                                     tweet.date,
                                     tweet.id,
                                     tweet.likeCount,
                                     tweet.viewCount,
                                     tweet.replyCount,
                                     tweet.retweetCount,
                                     tweet.sourceLabel,
                                     tweet.lang,
                                     tweet.retweetedTweet,
                                     tweet.quotedTweet,
                                     tweet.rawContent])
    
    # Mark progress
    index = user_names.index(user_name)
    if index % 100 == 0:
        end = time.perf_counter()
        elapsed = end - start
        print(f'Progress - {index} of {len(user_names)}',
              f'\nTime Elapsed - {elapsed:02.0f}:{elapsed % 1 * 60:02.0f}\n')
```

```python
# Creating a dataframe from the tweets list above 
tweets_df = pd.DataFrame(attributes_container, columns=["twitter_id",
                                                        "username",
                                                        "verified_status",
                                                        "date_created",
                                                        "tweet_id",
                                                        "likes",
                                                        "views",
                                                        "replies",
                                                        "retweets",
                                                        "source",
                                                        "language",
                                                        "retweeted_id",
                                                        "quoted_id",
                                                        "text"])

print(f'Result - Total count of tweets: {len(tweets_df)}')
```

```python
melted = (candidates
          .melt(id_vars = [x for x in candidates.columns if x not in ['twitter_name','user_name']],
                value_vars = ['twitter_name','user_name'],
                value_name = 'twitter_name')
          .drop(['variable','twitter_id'], axis = 1)
          .drop_duplicates()
          .sort_values('name')
          .reset_index(drop = True)
         )
```

```python
%%capture

# Created a list to capture all missing twitter accounts
exceptions2 = []
ids_from_names = []

for user_name in melted['twitter_name'].dropna().drop_duplicates():

    # Use ID to obtain username
    try:
        user_id = str(sntwitter.TwitterUserScraper(user_name).entity.id)
    except:
        exceptions2.append(user_name)
        continue
    
    # Add resulting user name to candidates df
    index = melted.index[melted['twitter_name'] == user_name]
    melted.loc[index, 'twitter_id'] = str(user_id)
    ids_from_names.append(user_id)
```

```python
pd.Series(exceptions2).drop_duplicates()
```

```python
melted['twitter_id'].dropna().drop_duplicates()
```

```python
melted['dupes'] = melted.duplicated('fec_id', keep = False)
melted = melted.query('dupes == False | ~twitter_name.isnull()').drop('dupes', axis = 1)
```

```python
melted['twitter_status_tweet'] = ''
melted.loc[melted['twitter_name'].isin(missing), 'twitter_status_tweet'] = 'No tweets found.'
melted['twitter_status_name'] = ''
melted.loc[melted['twitter_name'].isin(exceptions2), 'twitter_status_name'] = 'Name not found.'
melted['twitter_status_id'] = ''
melted.loc[melted['twitter_id'].isin(exceptions), 'twitter_status_id'] = 'ID not found.'
```
