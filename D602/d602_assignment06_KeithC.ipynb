# Dataset 1 - Weather

### Q1
Load your dataset into python using the pandas library and save data into a dataframe named dfi (where i is one of your datasets, for a total of 4).

```python
df1 = pd.read_csv('https://github.com/data602sps/datasetspractice/raw/main/weather.csv')
```

### Q2
Preview your data by calling your dataframe’s name. How many columns and rows do you see?

```python
df1
```

JupyterLab truncates the dataframe, showing the first 5 and last 5 rows. However, it does display all 15 columns.

### Q3
Examine the shape of your data using the .shape command and the data types of your columns using .dtypes.

```python
print(
    f'Shape: {df1.shape}\n'
    f'Data types: \n{df1.dtypes}'
)
```

The `shape` attribute returns the number of rows (8719) and columns (15). The `dtypes` attribute returns the data type of each column. Most columns are floats; the index and year columns are integers; and the origin column is an object (which as I understand, is a more efficient way of storing strings in pandas dataframes.

### Q4
Use .describe() on your data. What do you notice about your data? What does this command return?

```python
df1.describe()
```

The data is primarily numeric, as 14 of the 15 columns were returned by `.describe()`. There appear to be relatively few missing values, as most of the columns have the same count of 8718. It's also clear that all of the data is from year 2013.

### Q5
Use the .head() and .tail() command - what does this do?

```python
df1.head()
```

```python
df1.tail()
```

The `.head(n)` command returns the first $n$ rows, with $n=5$ as the default. The `.tail(n)` command returns the the *last* $n$ rows, with $n=5$ as the default.
