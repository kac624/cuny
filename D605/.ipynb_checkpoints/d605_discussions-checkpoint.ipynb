"""
In an upcoming national election for the President of the United States, a
pollster plans to predict the winner of the popular vote by taking a random
sample of 1000 voters and declaring that the winner will be the one obtaining
the most votes in his sample. Suppose that 48 percent of the voters plan
to vote for the Republican candidate and 52 percent plan to vote for the
Democratic candidate. To get some idea of how reasonable the pollster’s
plan is, write a program to make this prediction by simulation. Repeat the
simulation 100 times and see how many times the pollster’s prediction would
come true. Repeat your experiment, assuming now that 49 percent of the
population plan to vote for the Republican candidate; first with a sample of
1000 and then with a sample of 3000. (The Gallup Poll uses about 3000.)
(This idea is discussed further in Chapter 9, Section 9.1.)
"""

````python
n = 1000
p = 0.52

dem_votes = np.random.binomial(n, p)
rep_votes = n - dem_votes

print(
    f'Votes for Democrat: {dem_votes}\n'
    f'Votes for Republican: {rep_votes}'
)
````

$$ X \sim B(n,p) $$

````python
votes_df = pd.DataFrame(columns = ['dem_votes','rep_votes','proj_winner'])

n = 1000
p = 0.52

for i in range(100):

    dem_votes = np.random.binomial(n, p)
    rep_votes = n - dem_votes
    if dem_votes > rep_votes:
        proj_winner = 'dem'
    elif dem_votes < rep_votes:
        proj_winner = 'rep'
    else:
        proj_winner = 'tie'
    
    votes_df.loc[i,:] = [dem_votes, rep_votes, proj_winner]
    
print(
    'Proportion of Correct Predictions (i.e. Democrat Wins): '
    f'{votes_df.proj_winner.value_counts(normalize = True)[0]:.2%}\n'
)

fig, ax = plt.subplots()
ax.bar(votes_df.proj_winner.value_counts().index, 
       votes_df.proj_winner.value_counts().values)
plt.show()
````

There are n applicants for the director of computing. The applicants are interviewed independently by each member of the three-person search committee and ranked from 1 to n. A candidate will be hired if he or she is ranked first by at least two of the three interviewers. Find the probability that a candidate will be accepted if the members of the committee really have no ability at all to judge the candidates and just rank the candidates randomly. In particular, compare this probability for the case of three candidates and the case of ten candidates.

````python
import itertools
import numpy as np
import time
````

````python
start = time.time()

nb_case_hired = []
nb_combinations = []
prob_acceptance = []

for n in range(3,6):
    # Generate list of all possible rank permutations
    perms_n = list(itertools.permutations(np.arange(1,n+1)))
    # Generate all combinations of rankings produced by three committee members
    combos = itertools.combinations(perms_n * 3, 3)
    # Loop through each combination and count the number in which at least two members ranked the candidate in the top spot
    cases_hired = 0
    nb_combos_n = 0
    for combo in combos:
        nb_combos_n += 1
        nb_top_ranked = len([x for x in combo if x[0] == 1])
        if nb_top_ranked >= 2:
            cases_hired += 1
    # Calculate probability as cases in which the candidate is hired divided by total cases
    prob_n = cases_hired / nb_combos_n
    # Aggregate Results
    nb_case_hired.append(cases_hired)
    nb_combinations.append(nb_combos_n)
    prob_acceptance.append(prob_n)
    # Check time
    print(f'Done with n = {n}. Time: {(time.time() - start) / 60:.04f}.')

# Print results
print(
    '\n', nb_case_hired,'\n',
    nb_combinations,'\n',
    prob_acceptance
)
````

# Question

The lifetime, measure in hours, of the ACME super light bulb is a random variable $T$ with density function $f_T(t) = \lambda^{2}te^{-\lambda t}$, where $\lambda$ = 0.05. What is the expected lifetime of this light bulb? What is its variance?

# Post

The expected value and variance of a random variable $X$ with density function $f(x)$ is defined as follows (see Definition 6.4 and Theorem 6.13):
$$ E(X) = \int_{-\infty}^{\infty} xf(x)dx $$

$$ V(X) = \int_{-\infty}^{\infty} (x - \mu)^2 f(x) dx $$

In our case, $f(t)$ is given as $\lambda^{2}te^{-\lambda t}$, so we can plug that to these integrals.
$$ E(T) = \int_{-\infty}^{\infty} t * \lambda^{2}te^{-\lambda t} * dt $$

$$ V(T) = \int_{-\infty}^{\infty} (t - E(T))^2 * \lambda^{2}te^{-\lambda t} * dt $$

I won't solve this by hand (maybe after we get through the calculus portion of the course :D), but we can use Python's sympy package to evaluate the integrals. I'll begin by importing the package and defining the variables. We must set $t$ as a symbol so that we can later solve for it with sympy.

````python
from sympy import symbols, integrate, exp, oo

# Define variables
t = symbols('t')
lambda_ = 0.05
pdf = lambda_**2 * t * exp(-lambda_ * t)
````

According to the documentation for sympy's integrate function, we don't need to include the $dt$ in our call (i.e. "integrate(f, (x, a, b)) returns the definite integral $\int_{a}^{b} f dx$").

One thing I am not sure about is how to set $a$ and $b$ for $\int_{a}^{b}$. I believe in this case our random variable is bounded by 0 and $\infty$, since the lightbulb cannot have a lifetime less than zero, and it has no clear upper bound. So I'll set $a = 0$ and $b = \infty$ (we use oo as infinity in sympy).

With that, we simply spell out the formulas for each, then evaluate using the .evalf() method.

````python
# Formula for the expected value of T
E_T = integrate(t * pdf, (t, 0, oo))

# Formula for the expected value of T**2
V_T = integrate((t - E_T)**2 * pdf, (t, 0, oo))

# Evaluate the integrals and print the results
print(f'E(T): {E_T.evalf():.2f}')
print(f'V(T): {V_T.evalf():.2f}')
````

So, our expected liftime is 40 hours, with a variance of 800 hours (or a standard deviation of ~28.28).

Assume that the service time for a customer at a bank is exponentially distributed with mean service time 2 minutes. Let X be the total service time for 10 customers. Estimate the probability that X > 22 minutes.

## KC Response

Question

Assume that the service time for a customer at a bank is exponentially distributed with mean service time 2 minutes. Let X be the total service time for 10 customers. Estimate the probability that X > 22 minutes.

Response

I'm struggling quite a bit with this chapter, so any input is appreciated! I had to rely on lots of googling. While the textbook clearly covers the case of summing 2 exponential variables, I wasn't able to identify an example with more than 2. And in this question, we need to sum 10.

It seems that the key insight to solve this problem is that "the sum of $n$ exponential ($\beta$) random variables is a gamma ($n$, $\beta$) random variable. Since $n$ is an integer, the gamma distribution is also an Erlang distribution" [1]; also see [2]. The text only mentions the gamma distribution tangentially and doesn't seem to cover Erlang at all. But I gather from wiki (and a bunch of stack exchange responses) that you can define the PDF of an Erlang distribution as follows [3].

$f(x; k, \lambda) = \frac{\lambda^k x^{k-1} e^{-\lambda x}}{(k-1)!}$ for $x, \lambda \geq 0$

We're looking for the probability that X > 22, i.e. $P(X>22)$. We can use the compliment of the CDF to find this, i.e. $1-P(X \leq 22)$. Because $X$ is a continuous random variable, we can take the integral of PDF above to find this CDF. So, let's define the key parameters and solve!

The $k$ refers to a shape parameter, which I understand to be the number of random variables we're summing (i.e. 10). And $\lambda$ is the rate parameter, which would be 2 for a Poisson, but is the inverse for an exponential distribution, or 1/2. Finally, for our integral, we'll define the bounds as 0 and 22, since our variable is lower bound by zero (per the definition above), and we're looking for $P(X \leq 22)$, giving 22 as the upper bound.

We'll use Python's sympy package to solve the integral. Note that we then take it's complement to find $P(X>22)$.

```python
from sympy import symbols, integrate, exp, factorial

lambda_ = 1/2
k = 10
x = symbols('x')

erlang_pdf = lambda_**k * x**(k-1) * exp(-lambda_*x) / factorial(k-1)

p_Xgreater22 = integrate(erlang_pdf, (x, 0, 22))

print(f'P(X>22) = {1 - round(p_Xgreater22,5)}')
```
\>\>\> P(X>22) = 0.34051

So, the probability that the total service time for 10 customers is less than 22 is ~0.34051.

Python also provides a way to directly calculate the CDF of an erlang distribution via the scipy.stats package. From the scipy documentation [4], the erlang() method takes three arguments: x, a, and scale. The x argument is our $x$ value (22); the a argument is our shape parameter, or the $k$ value (10); and the scale argument is our mean (2).

```python
import scipy.stats as sps

erlang_cdf = sps.erlang.cdf(22, a = 10, scale = 2)

print(f'P(X>22) = {1 - round(erlang_cdf,5)}')
```
\>\>\> P(X>22) = 0.34051

We get the same result!

We have two coins: one is a fair coin and the other is a coin that produces heads with probability 3/4. One of the two coins is picked at random, and this coin is tossed n times. Let Sn be the number of heads that turns up in these n tosses. Does the Law of Large Numbers allow us to predict the proportion of heads that will turn up in the long run? After we have observed a large number of tosses, can we tell which coin was chosen? How many tosses suffice to make us 95 percent sure?

-----------

Great response. I like the theoretical approach. I'm also unsure how to directly determine the sample size needed to make a 95% confidence conclusion. I believe it would involve a power test (similar to problems we did in D606 related to the required sample size to obtain a certain margin of error).

As an alternative, I set up a simulation to test results using various sample sizes. The simulation randomly chooses one of the two coins and, using associated the probability for heads, conducts a bunch of coin flips. Based on that sample, it then concludes whether the coin is fair or biased based on a 95% confidence one-proportion z-test ($H_0: p = p_0$; $H_A: p \neq p_0$). The same procedure is then repeated for the other coin.

The simulation indicates that, even with sample sizes as small as 100, we can conclude with 95% which coin is fair and which is biased. With a sample size of only 10, however, the results are inconclusive. The mean would imply that coin 1 is fair and coin 2 is biased, but we can't conclude that with 95% confidence.

````python
import numpy as np
from statsmodels.stats.proportion import proportions_ztest
np.random.seed(42)

# setup parameters
probs = [0.5, 0.75]
n_sims = [10**1, 10**2, 10**3, 10**4, 10**5]

# simulate using various sample sizes
for n in n_sims:
    # randomly sample one of two coins, flip n times, calculate p-value
    prob1 = probs[np.random.choice([0,1])]
    sample1 = np.random.choice([1,0], p = [prob1, 1-prob1], size = n)
    pval1 = proportions_ztest(
        count = len(sample1[sample1 == 1]),
        nobs = len(sample1),
        alternative = 'two-sided',
        value = 0.5, prop_var = 0.5
    )
    conclusion1 = 'fair' if pval1[1] <= 0.025 else 'biased'
    
    # repeat for second coin, using whichever probability wasn't used in the first
    prob2 = [x for x in probs if x != prob1][0]
    sample2 = np.random.choice([1,0], p = [prob2, 1-prob2], size = n)
    pval2 = proportions_ztest(
        count = len(sample2[sample2 == 1]),
        nobs = len(sample2),
        alternative = 'two-sided',
        value = 0.5, prop_var = 0.5
    )
    conclusion2 = 'fair' if pval2[1] <= 0.025 else 'biased'
    
    # print results
    print(
        f'With n = {n:,}\n\n'
        
        f'Mean of sample1: {sample1.mean():.2f}\n'
        f'P-value for sample1: {pval1[1]:.5f}\n'
        f'95% conclusion: {conclusion1}\n\n'
        
        f'Mean of sample2: {sample2.mean():.2f}\n'
        f'P-value for sample2: {pval2[1]:.5f}\n'
        f'95% conclusion: {conclusion2}\n-------\n'
    )
````

````python
# we have three subsets, based on how the y and X variables are transformed
growth_cols = [x for x in df.columns if 'growth' in x]
diff_cols = [x for x in df.columns if 'diff' in x]
level_cols = [x for x in df.columns if 'growth' not in x and 'diff' not in x]

variable_groups = [growth_cols, diff_cols, level_cols]
variable_group_names = ['growth', 'diff', 'level']
number_features = 3

# loop through each subset and check each combination of variables
# If a combo has more than one type of the sample variable, remove that combo
# For example, if a combo has gdp_lag and gdp_lead, remove it
for group, group_name in zip(variable_groups, variable_group_names):
    subset = df[group]
    variables = [x for x in subset.columns if 'earnings' not in x]
    all_combos = itertools.combinations(variables, number_features)
    globals()['var_combos_' + group_name] = []

    start = time.perf_counter()
    for i, combo in enumerate(all_combos):

        exclude = False
        for series in series_ids.keys():
            if ' '.join(combo).count(series) > 1:
                exclude = True
        if exclude == False:
            globals()['var_combos_' + group_name].append(combo)

    print(f'Final no. of combinations for {group_name}: {len(globals()["var_combos_" + group_name]):,}')
````

````python
%%capture --no-stdout

starting_length = len(model_fits)
start = time.perf_counter()

# include only models that overall appear significant based on F-test
model_fits_clean = {key: value for key, value in model_fits.items() if value.f_pvalue <= 0.1}
print(f'Models removed due to significance: {starting_length - len(model_fits_clean):,}')

# include only models with ~normal residuals
starting_length = len(model_fits_clean)
model_fits_clean = {
    key: value for key, value in model_fits_clean.items() 
    if sps.shapiro(value.resid)[1] >= 0.001
}
print(f'Models removed due to non-normal residuals: {starting_length - len(model_fits_clean):,}')

# include only models with stationary residuals
starting_length = len(model_fits_clean)
model_fits_clean = {
    key: value for key, value in model_fits_clean.items() 
    if sm.tsa.stattools.adfuller(value.resid)[1] >= 0.05
}
print(f'Models removed due to non-stationary residuals: {starting_length - len(model_fits_clean):,}')

# include only models with max VIF lower than 10
starting_length = len(model_fits_clean)
model_fits_clean = {
    key: value for key, value in model_fits_clean.items() 
    if max([variance_inflation_factor(value.model.exog, i) for i in range(number_features+1)]) >= 10
}
print(f'Models removed due to multicollinearity: {starting_length - len(model_fits_clean):,}')

now = time.perf_counter()
print(
    f'Remaining number of models: {len(model_fits_clean):,}\n',
    f'Time to complete: {(now-start)/60:.2f} minutes'
)
````

````python
# Use dict comprehension to pull out key values from model fits
model_data = {
    key: [
        value.params.index.tolist(),
        value.params.tolist(),
        value.rsquared_adj,
        value.mse_resid,
        value.aic
    ] for key, value in model_fits_clean.items()
}

# Convert dict to dataframe, assign columns
model_data = pd.DataFrame.from_dict(
    model_data, orient = 'index',
    columns = [
        'predictors',
        'coefficients',
        'r2',
        'mse',
        'aic'
    ]
)

model_data.head()
````

````python
# The orientation of r2 (higher is better) is opposite of mse and aic (lower is better)
# So I create an inverse r2 column and use that in the ranking
model_data['r2_inverse'] = 1 - model_data.r2
model_data['rank'] = model_data.loc[:,['mse','aic','r2_inverse']].rank(method = 'first').mean(axis = 1)
model_data.drop('r2_inverse', axis = 1, inplace = True)
model_data.sort_values('rank', inplace = True)

# Let's see the top 10 best and worst
display(
    model_data.head(10),
    model_data.tail(10)
)
````

````python
best_model = model_fits_clean[model_data.index[0]]

print(best_model.summary())

fig, ax = plt.subplots(figsize=(14, 8)) 
ax.scatter(best_model.fittedvalues, best_model.resid)
plt.show()
````
